```python
# ÌÅ¨Î°§ÎßÅÌïú ÏóëÏÖÄ ÌååÏùº ÏùΩÏñ¥Ïò§Í∏∞
df = pd.read_excel("Youtube.xlsx")
```

```python
### Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ 

## DriveÎûÄ Îã®Ïñ¥Îßå Îì§Ïñ¥Í∞ÄÎäî ÌÖçÏä§Ìä∏ Ï∞æÍ∏∞ 
drive = df[df["Ï†úÎ™©"].str.contains("Drive")] 

## 4KÎûÄ Îã®Ïñ¥Îßå Îì§Ïñ¥Í∞ÄÎäî ÌÖçÏä§Ìä∏ Ï∞æÍ∏∞
drive_4K = drive[drive["Ï†úÎ™©"].str.contains("4K")] 

# Ïù∏Îç±Ïä§ Ïû¨Ï†ïÎ†¨
drive_new_line = drive_4K.reset_index(drop=True) 
drive_new_line
```

```python
# Ï∂îÏ∂úÌï† Ïù∏Îç±Ïä§Îßå Í∞ÄÏ†∏Ïò§Í∏∞ 
national = drive_new_line.iloc[[0,2,4,7,9,11,19,22,30,33,34,40,44,66,79,101,102,104,107,109,110,137]] 

# Ïù∏Îç±Ïä§ Ïû¨Ï†ïÎ†¨
national.reset_index(drop=True) 
```

```python
# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê ÎÑ£ÏùÑ Í∞í ÏÑ§Ï†ï
continents = ['Monaco']+ ['Wales'] + ['Ireland'] + ['Bristol, England'] +['South Africa'] + ['Mumbai, India'] + ['New York, USA'] +['Dubai, UAE']\
+ ['Los Angeles, USA'] + ['Amsterdam, Netherlands'] + ['Madrid, Spain'] + ['Rome, Italy'] + ['Zurich, Switzerland'] + ['Istanbul, Turkey']\
+ ['Cairo, Egypt'] + ['Buenos Aires, Brazil'] + ['Hong Kong'] + ['Singapore'] + ['Santiago, Chile'] + ['Tokyo, Japan'] + ['Toronto, Canada'] + ['Paris, France']

# ÏßÄÏó≠/Íµ≠Í∞ÄÎ™Ö Ïª¨Îüº Ï∂îÍ∞Ä
national['ÏßÄÏó≠/Íµ≠Í∞ÄÎ™Ö'] = continents 

# ÏßÄÏó≠/Íµ≠Í∞ÄÎ™Ö Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨ 
final_nation = national.sort_values(by="ÏßÄÏó≠/Íµ≠Í∞ÄÎ™Ö").reset_index(drop=True)  
final_nation    
```
